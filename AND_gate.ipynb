```python
import numpy as np
np.random.seed(42)
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
    
```

```python
## AND Dataset
```

```python
X=np.array([[0,0],[0,1],[1,0],[1,1]])
y=np.array([[0],[0],[0],[1]])
print(X)
print(y)
```

```python
W1=np.random.randn(2,4)
b1=np.zeros((1,4))
W2=np.random.randn(4,1)
b2=np.zeros((1,1))
print(W1.shape,W2.shape)
```

```python
z1=X@W1+b1
a1=sigmoid(z1)
z2=a1@W2+b2
prediction=sigmoid(z2)
print(prediction)
```

```python
error=prediction-y
print(error)
```

```python
delta2=error*prediction*(1-prediction)
delta1=(delta2@W2.T)*a1*(1-a1)
print(delta2)
print(delta1)
```

```python
lr=0.1
W2-=lr*(a1.T@delta2)
b2-=lr*np.sum(delta2,axis=0,keepdims=True)
W1-=lr*(X.T@delta1)
b1-=lr*np.sum(delta1,axis=0,keepdims=True)
print("Weights Updated!")
```
