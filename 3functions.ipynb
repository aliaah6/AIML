```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-10,10,100)

sigmoid=1/(1+np.exp(-x))

plt.plot(x,sigmoid)
plt.title("Step Activation Function")
plt.grid()
plt.show()
```

```python
x=np.linspace(-10,10,100)

tanh=np.tanh(x)
plt.plot(x,tanh)
plt.title("Tan Activation Function")
plt.grid()
plt.show()
```

```python
x=np.linspace(-10,10,100)

relu=np.maximum(0,x)
plt.plot(x,relu)
plt.title("RELU Activation Function")
plt.grid()
plt.show()
```

```python
import math

def step_function(z):
    if z >= 0:
        return 1
    return 0

def tanh_function(z):
    return (math.exp(z) - math.exp(-z)) / (math.exp(z) + math.exp(-z))

def relu(z):
    return max(0, z)
# 1. Step Function - Spam Email Detection

def spam_detector(links, spam_words, unknown_sender):
    weights = [0.4, 0.4, 0.3]
    bias = -0.5

    z = links*weights[0] + spam_words*weights[1] + unknown_sender*weights[2] + bias
    output = step_function(z)
    return z, output

print("\n========== STEP FUNCTION : SPAM DETECTION ==========\n")

emails = [
    ("Normal Email", 0.1, 0.1, 0),
    ("Promotional Email", 0.5, 0.4, 1),
    ("Spam Email", 0.9, 0.8, 1)
]

for name, links, spam_words, sender in emails:
    z, out = spam_detector(links, spam_words, sender)

    print("Email:", name)
    print("Weighted Sum (z):", round(z,3))
    print("Spam" if out==1 else "Not Spam")
    print("-"*35)

# 2. Tanh Function - Product Review Sentiment

def sentiment(positive, negative, neutral):
    weights = [0.8, -0.9, -0.2]
    bias = 0

    z = positive*weights[0] + negative*weights[1] + neutral*weights[2] + bias
    output = tanh_function(z)
    return z, output

print("\n========== TANH FUNCTION : PRODUCT REVIEW ==========\n")

reviews = [
    ("Excellent Product", 5, 0, 0),
    ("Average Product", 2, 2, 1),
    ("Poor Product", 0, 5, 0)
]

for name, pos, neg, neu in reviews:
    z, out = sentiment(pos, neg, neu)

    print("Review:", name)
    print("Weighted Sum (z):", round(z,3))
    print("Tanh Output:", round(out,3))
    print("-"*35)

# 3. ReLU Function - Student Performance Score

def student_score(attendance, assignment, exam):
    weights = [0.2, 0.3, 0.5]
    bias = -0.2

    z = attendance*weights[0] + assignment*weights[1] + exam*weights[2] + bias
    output = relu(z)
    return z, output

print("\n========== ReLU FUNCTION : STUDENT PERFORMANCE ==========\n")

students = [
    ("Top Student", 0.95, 0.90, 0.92),
    ("Average Student", 0.65, 0.60, 0.58),
    ("Weak Student", 0.30, 0.25, 0.20)
]

for name, attendance, assignment, exam in students:
    z, out = student_score(attendance, assignment, exam)

    print("Student:", name)
    print("Weighted Sum (z):", round(z,3))
    print("ReLU Output:", round(out,3))
    print("-"*35)
```
