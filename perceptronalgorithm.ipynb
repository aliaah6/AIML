```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.linear_model import Perceptron
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
x=digits.data
y=digits.target

y=np.array([0 if label%2==0 else 1 for label in y])

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

model=Perceptron(max_iter=1000)

model.fit(x_train,y_train)

y_pred=model.predict(x_test)

accuracy=print("Accuracy:\n",accuracy_score(y_test,y_pred))
confusion=print("Confusion matrix:\n",confusion_matrix(y_test,y_pred))

for i in range(9):
    plt.imshow(x_test[i].reshape(8,8),cmap='gray')

    actual="Even" if y_test[i]==0 else "odd"
    predicted="Even" if y_test[i]==0 else "odd"

    plt.title(f"actual: {actual} predicted:{predicted}")
    plt.axis("off")
    plt.show()
```
