```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("All libraries imported successfully!")
```

```python
df=pd.read_csv("Salary_Data.csv")
```

```python
plt.scatter(df["YearsExperience"],df["Salary"],label="Salary")

#add the chart with info to make it more readable
plt.title("Experience vs Salary")
plt.xlabel("YearsExperience")
plt.ylabel("Salary")

#show the legend
plt.legend()

plt.grid(axis="y",alpha=0.4)

sns.pairplot(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("Salary_Data.csv")

# Plot Salary only
plt.plot(df["YearsExperience"], df["Salary"], label="Salary", color="green")
plt.scatter(
    df["YearsExperience"], df["Salary"], label="Salary", color="green"
)

# Chart Information
plt.title("Experience VS Salaries")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()  # Added to display the "Salary" label
plt.grid(axis="y", alpha=0.4)

# Show pairplot
sns.pairplot(df)
plt.show()
```
