```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python

# ---------------- Dataset ----------------
data = {
    "Experience": [2, 4, 6, 8, 10],
    "Salary": [25000, 35000, 50000, 65000, 80000]
}

df = pd.DataFrame(data)

# ---------------- Train Model ----------------
X = df[["Experience"]]
y = df["Salary"]

model = LinearRegression()
model.fit(X, y)

# ---------------- User Input ----------------
experience = float(input("Enter Experience (Years): "))

# ---------------- Prediction ----------------
predicted_salary = model.predict([[experience]])[0]

print(f"\nPredicted Salary = ₹{predicted_salary:,.2f}")

# ---------------- Plot ----------------
plt.figure(figsize=(8,5))

# Actual data points
plt.scatter(
    df["Experience"],
    df["Salary"],
    color="blue",
    s=80,
    label="Training Data"
)

# Regression line (extend to entered experience)
max_x = max(experience, df["Experience"].max())

x_line = np.linspace(0, max_x, 200).reshape(-1, 1)
y_line = model.predict(x_line)

plt.plot(
    x_line,
    y_line,
    color="red",
    linewidth=3,
    label="Regression Line"
)

# Predicted point
plt.scatter(
    experience,
    predicted_salary,
    color="green",
    marker="*",
    s=250,
    label="Predicted Salary"
)

# Display prediction text
plt.text(
    experience,
    predicted_salary,
    f" ₹{predicted_salary:,.0f}",
    fontsize=10
)

# Extend x-axis to the entered value
plt.xlim(0, max_x + 1)

plt.title("Linear Regression: Experience vs Salary")
plt.xlabel("Experience (Years)")
plt.ylabel("Salary (₹)")
plt.grid(True)
plt.legend()

plt.show()


# ---------------- Pair Plot ----------------
sns.pairplot(df)
plt.show()
```
